```r
library(dplyr)
library(ggplot2)
library(boot)
library(glmnet)
```

```r
train <- read.csv("https://raw.githubusercontent.com/ajschumacher/gadsdata/master/salary/train.csv",
                  stringsAsFactors = FALSE)
head(train, 5)
```

```r
options(repr.plot.width = 6)
options(repr.plot.height = 4)

hist(train$SalaryNormalized, breaks = 100, col = "lavender", xlab = "Salary") + 
    abline(v = mean(train$SalaryNormalized), lwd = 3)
```

```r
apply(train, 2, function(x) {length(unique(x))}) #CHECK THE UNIQUE VALUES.
2+2
```

```r
Now let's look at the most popular job titles and what the mean salary for each of these is:
```

```r
train %>% 
    group_by(Title) %>% 
    summarize( count = n(),
               mean = mean(SalaryNormalized)) %>% 
    filter(count >10) %>% 
    arrange(desc(count)) %>% head(10)
```

```r
train$Title[train$Title == ""] <- "_" #otherwise it messes up

#you start with an object that you want to apply a bunch of functions not not with the functions themselves:
train$Title %>% sapply(function(x) {strsplit(x, " ")}[[1]] %>% tail(1)) -> train$generic

train$generic[1:10]
```

```r
train %>%
    count(generic)  %>%
    arrange(desc(n)) %>%
    head(10) -> toptitles
```

```r
ggplot(toptitles, aes(x = reorder(generic, n), y = n)) +
    geom_bar(stat = "identity") +
    coord_flip()
```

```r
barplot(toptitles$n, names.arg = toptitles$generic, horiz = TRUE, las =1)
```
